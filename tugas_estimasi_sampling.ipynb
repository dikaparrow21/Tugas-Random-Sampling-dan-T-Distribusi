```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Tambahkan huruf r sebelum tanda kutip pembuka
data = pd.read_csv(r'C:\Users\asus\OneDrive\Documents\statistika\Estimasi\MovieData-2018-2020.csv')

data.head(30)
```

```python
data['genre'].value_counts()
```

```python
#hilangkan genre yang hanya berisi 1 judul
filter_data = data.groupby('genre').filter(lambda x: len(x) > 1)
filter_data['genre'].value_counts()
```

```python
_, stratified_sample = train_test_split(
    filter_data,
    test_size=0.2,
    stratify=filter_data['genre'],
    random_state=53
)
sample = stratified_sample.head(30)
data_budget = sample['budget']

print("Stratified Sampling (30 data teratas):\n", sample)
```

```python
import scipy.stats as stats
import math

x_bar = 6.48 #rata-rata
s = 7.57 # std sample (standar deviasi)
n = 30 # jumlah data(sampel)
confidence_level = 0.90 # pilih antara 0.90 / 0.95 / 0.99

alpha = (1 - confidence_level) # mencari alpha
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha/2, df)
margin_of_error = critical_t * standard_error

lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

print("Critical t-value: {:.2f}".format(critical_t))
print("Margin of Error: {:.2f}".format(margin_of_error))
print("Confidence Interval: [{:.2f},{:.2f}]".format(lower_bound, upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.2f} and {:.2f}".format(lower_bound, upper_bound))
```
